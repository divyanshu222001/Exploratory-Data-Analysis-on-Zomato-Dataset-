```python
#Creating a pie chart to identify the top three countries with the highest Zomato usage.
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

```python
city_values = final_df['City'].value_counts().values
city_indices = final_df['City'].value_counts().index
```

```python


# Take the top 5 cities
top_cities_values = city_values[:5]
top_cities_labels = city_indices[:5]

```

```python
# Plot the pie chart
plt.pie(top_cities_values, labels=top_cities_labels, autopct='%1.1f%%')
plt.title('Top 5 Cities on Zomato')  # Add a title if needed
plt.show()
```

```python
# Assuming 'Cuisine' is the column containing cuisine information
top_cuisine_counts = final_df['Cuisines'].value_counts().head(10)
```

```python

# Extracting the top 10 cuisines and their counts
top_cuisine_values = top_cuisine_counts.values
top_cuisine_labels = top_cuisine_counts.index
```

```python
# Plotting a bar chart for better visualization
plt.figure(figsize=(10, 6))
plt.bar(top_cuisine_labels, top_cuisine_values, color='skyblue')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title('Top 10 Cuisines on Zomato')
plt.xticks(rotation=45, ha='right')
plt.show()
```
